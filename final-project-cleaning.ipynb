```python
import re
import pandas as pd
```

```python
#Bring in all data files.
location_person = '/Users/ebrushka/Desktop/python/Final Project/location bus:home etc/Crimes_Against_Persons_Offenses_Offense_Category_by_Location_2022.xlsx'
location_property = '/Users/ebrushka/Desktop/python/Final Project/location bus:home etc/Crimes_Against_Property_Offenses_Offense_Category_by_Location_2022.xlsx'
location_society = '/Users/ebrushka/Desktop/python/Final Project/location bus:home etc/Crimes_Against_Society_Offenses_Offense_Category_by_Location_2022.xlsx'

time_person = '/Users/ebrushka/Desktop/python/Final Project/timeOfDay/Crimes_Against_Persons_Incidents_Offense_Category_by_Time_of_Day_2022.xlsx'
time_property ='/Users/ebrushka/Desktop/python/Final Project/timeOfDay/Crimes_Against_Property_Incidents_Offense_Category_by_Time_of_Day_2022.xlsx'
time_society = '/Users/ebrushka/Desktop/python/Final Project/timeOfDay/Crimes_Against_Society_Incidents_Offense_Category_by_Time_of_Day_2022.xlsx'

victims_age = '/Users/ebrushka/Desktop/python/Final Project/victims/Victims_Age_by_Offense_Category_2022.xlsx'
victims_race = '/Users/ebrushka/Desktop/python/Final Project/victims/Victims_Race_by_Offense_Category_2022.xlsx'
victims_sex = '/Users/ebrushka/Desktop/python/Final Project/victims/Victims_Sex_by_Offense_Category_2022.xlsx'

agencies = '/Users/ebrushka/Desktop/python/Final Project/agencies.csv'
offense_by_agency = '/Users/ebrushka/Desktop/python/Final Project/New_Jersey_Offense_Type_by_Agency_2022.xlsx'
```

```python
#Read each Excel file into a Pandas DataFrame. To ensure data comes inclean, skip unnecessary rows.
df_location_person = pd.read_excel(location_person,skiprows=4)
df_location_property = pd.read_excel(location_property,skiprows=4)
df_location_society = pd.read_excel(location_society,skiprows=4)

df_time_person = pd.read_excel(time_person,skiprows=4)
df_time_property = pd.read_excel(time_property,skiprows=4)
df_time_society = pd.read_excel(time_society,skiprows=4)

df_victims_age = pd.read_excel(victims_age,skiprows=4)
df_victims_race = pd.read_excel(victims_race,skiprows=4)
df_victims_sex = pd.read_excel(victims_sex,skiprows=4)

df_agencies = pd.read_csv(agencies)
df_offense_by_agency = pd.read_excel(offense_by_agency,skiprows=4)
```

### Clean the column names and remove information unnesecary for our analysis. 

```python
#Since we skipped the unnecesarry rows, we lost two of the headers because of the format. Let's manually put those in.
df_location_person = df_location_person.rename(columns={df_location_person.columns[0]: 'Location', df_location_person.columns[1]: 'Total_person'})

#Clean out the newline and whitespace in the column names
cleaned_columns = [re.sub(r'[\n\s]+', '', col) for col in df_location_person.columns]
df_location_person.columns = cleaned_columns

df_location_person
```

```python
#Manually input some column names. Clean out the newline and whitespace in the column names.
df_location_property = df_location_property.rename(columns={df_location_property.columns[0]: 'Location', df_location_property.columns[1]: 'Total_property'})
cleaned_columns = [re.sub(r'[\n\s]+', '', col) for col in df_location_property.columns]
df_location_property.columns = cleaned_columns
df_location_property
```

```python
#Manually input some column names. Clean out the newline and whitespace in the column names.
df_location_society = df_location_society.rename(columns={df_location_society.columns[0]: 'Location', df_location_society.columns[1]: 'Total_society'})
cleaned_columns = [re.sub(r'[\n\s]+', '', col) for col in df_location_society.columns]
df_location_society.columns = cleaned_columns
df_location_society
```

```python
#Manually input some column names. Clean out the newline and whitespace in the column names.
df_time_person = df_time_person.rename(columns={df_time_person.columns[0]: 'Time', df_time_person.columns[1]: 'Total_person'})
cleaned_columns = [re.sub(r'[\n\s]+', '', col) for col in df_time_person.columns]
df_time_person.columns = cleaned_columns
#Remove unneessary row.
df_time_person = df_time_person[:-1]
df_time_person
```

```python
#Manually input some column names. Clean out the newline and whitespace in the column names.
df_time_property = df_time_property.rename(columns={df_time_property.columns[0]: 'Time', df_time_property.columns[1]: 'Total_property'})
cleaned_columns = [re.sub(r'[\n\s]+', '', col) for col in df_time_property.columns]
df_time_property.columns = cleaned_columns
#Remove unnecessary row.
df_time_property = df_time_property[:-1]
df_time_property
```

```python
#Manually input some column names. Clean out the newline and whitespace in the column names.
df_time_society = df_time_society.rename(columns={df_time_society.columns[0]: 'Time', df_time_society.columns[1]: 'Total_society'})
cleaned_columns = [re.sub(r'[\n\s]+', '', col) for col in df_time_society.columns]
df_time_society.columns = cleaned_columns
#Remove unnecessary row.
df_time_society = df_time_society[:-1]
df_time_society
```

```python
#Manually input some column names. Clean out the newline and whitespace in the column names.
df_victims_age = df_victims_age.rename(columns={df_victims_age.columns[0]: 'Category', df_victims_age.columns[1]: 'Total_age'})
cleaned_columns = [re.sub(r'[\n\s]+', '', col) for col in df_victims_age.columns]
df_victims_age.columns = cleaned_columns
#Remove unnecessary row.
df_victims_age = df_victims_age[:-1]
df_victims_age
#It looks like it will be best to categorize some of the ages together but we can do that according to the 
#distribution later in the analysis.
```

```python
#Manually input some column names. Clean out the newline and whitespace in the column names.
df_victims_race = df_victims_race.rename(columns={df_victims_race.columns[0]: 'Category', df_victims_race.columns[1]: 'Total_race'})
cleaned_columns = [re.sub(r'[\n\s]+', '', col) for col in df_victims_race.columns]
df_victims_race.columns = cleaned_columns
#Remove unnecessary row.
df_victims_race = df_victims_race[:-1]
df_victims_race
```

```python
#Manually input some column names. Clean out the newline and whitespace in the column names.
df_victims_sex = df_victims_sex.rename(columns={df_victims_sex.columns[0]: 'Category', df_victims_sex.columns[1]: 'Total_sex'})
cleaned_columns = [re.sub(r'[\n\s]+', '', col) for col in df_victims_sex.columns]
df_victims_sex.columns = cleaned_columns
#Remove unnecessary row.
df_victims_sex = df_victims_sex[:-1]
df_victims_sex
```

```python
#Remove columns we do not need.
columns_to_remove = [
    'data_year', 'parent_pop_group_desc', 'population_group_desc',
    'suburban_area_flag', 'state_postal_abbr', 'submitting_agency_name',
    'population', 'region_name'
]

df_agencies = df_agencies.drop(columns=columns_to_remove)
df_agencies
```

```python
#Manually input some column names. Clean out the newline and whitespace in the column names.
cleaned_columns = [re.sub(r'[\n\s]+', '', col) for col in df_offense_by_agency.columns]
df_offense_by_agency.columns = cleaned_columns
#Remove unnecessary row.
df_offense_by_agency = df_offense_by_agency[:-1]
#Drop unnecessary column due to formatting.
df_offense_by_agency = df_offense_by_agency.drop(df_offense_by_agency.columns[0], axis=1)
#Drop unnecessary column.
df_offense_by_agency = df_offense_by_agency.drop(columns='Population2')
df_offense_by_agency
```

### Join data that match up together to have less dataframes.

```python
#Let's merge the offenses and the agencies dataframes to be able to answer where the highest crimes are.
merged_county_df = pd.merge(df_offense_by_agency, df_agencies[['pub_agency_name', 'county_name']], left_on='AgencyName', right_on='pub_agency_name', how='left')
#Drop the redundant column 'pub_agency_name' if needed
merged_county_df = merged_county_df.drop(columns=['pub_agency_name'])
merged_county_df
```

```python
#Let's make sure we can get the total number of offenses in county so we can later visualize and analyze it.
total_offenses_by_county = merged_county_df.groupby('county_name')['TotalOffenses'].sum().reset_index()
total_offenses_by_county
#Success!!
```

```python
#Let's merge all the location data to be in one dataframe.
merged_location_df = pd.merge(df_location_person, df_location_property, on='Location', how='outer', suffixes=('_person', '_property'))
merged_location_df = pd.merge(merged_location_df, df_location_society, on='Location', how='outer', suffixes=('', '_society'))
merged_location_df
```

```python
#Let's merge all the time data to be in one dataframe.
merged_time_df = pd.merge(df_time_person, df_time_property, on='Time', how='outer', suffixes=('_person', '_property'))
merged_time_df = pd.merge(merged_time_df, df_time_society, on='Time', how='outer', suffixes=('', '_society'))
merged_time_df
```

```python
#Let's merge all the victim data to be in one dataframe.
merged_victims_df = pd.merge(df_victims_age, df_victims_race, on='Category')
merged_victims_df = pd.merge(merged_victims_df, df_victims_sex)
merged_victims_df
```

### Let's save our new dataframes into CSV files.
